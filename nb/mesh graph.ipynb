```python
# download latest
url = "https://www.nlm.nih.gov/mesh/2019/download/2019New_Mesh_Tree_Hierarchy.txt"

import requests, shutil

r = requests.get((url), verify=True,stream=True)
r.raw.decode_content = True
with open("data/mesh_tree.txt", 'wb') as f:
    shutil.copyfileobj(r.raw, f) 
```

```python
import re
from collections import defaultdict

name_from_id = {}
subtrees = defaultdict(list)

# pass 1

with open("data/mesh_tree.txt", 'r') as f:
    for i, l in enumerate(f):
        if i<2 or not l.strip() or l.startswith('---'):
            continue
            
        parts = re.split('\s+', l)
        code = parts[0]
        
        if code=='TREE_NUMBER':
            continue
        
        desc = ' '.join(parts[1:])

        name_from_id[code] = desc.strip()
        code_parts = code.split('.')
        
        if len(code_parts) > 1:
            subtrees[name_from_id['.'.join(code_parts[:-1])]].append(name_from_id[code])
        
#         for k in range(1, len(code_parts)+1):
#             subtrees_c['.'.join(code_parts[:k])].append(code)
            
            
            
#         pos = tree['root'] 
        # as changes to pos will change the underlying tree
        
#         for p in code_parts:
#             if p not in pos:
#                 pos[p] = {}
#             
```
